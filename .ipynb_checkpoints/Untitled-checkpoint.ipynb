```python
print("📥 Loading Dataset...")

df_sh = pd.read_csv('Car details v3.csv')
df_sh['brand'] = df_sh['name'].str.split().str[0]
df_sh.rename(columns={'selling_price': 'second_hand_price'}, inplace=True)

print(df_sh.shape)
df_sh.head()
```

```python
msrp_data = {
    'Maruti': {'Swift': 850000, 'Alto': 350000, 'Baleno': 650000, 'Brezza': 850000},
    'Hyundai': {'Creta': 1450000, 'Venue': 850000, 'i20': 750000, 'Verna': 1050000},
    'Toyota': {'Innova': 1950000, 'Fortuner': 3250000, 'Corolla': 1850000},
    'BMW': {'X5': 8500000, '3 Series': 4500000, '5 Series': 6500000, 'X3': 5500000},
    'Mercedes': {'E-Class': 6500000, 'C-Class': 4500000, 'GLA': 4500000},
    'Audi': {'Q7': 8500000, 'A4': 4500000, 'Q5': 5500000},
}

msrp_rows = []

for brand, models in msrp_data.items():
    for model, price in models.items():
        msrp_rows.append({
            'brand': brand,
            'name': model,
            'original_msrp': price,
            'year': 2024,
            'km_driven': 0,
            'type': 'new_msrp'
        })

df_msrp = pd.DataFrame(msrp_rows)

df = pd.concat([df_sh, df_msrp], ignore_index=True)

df.shape
```

```python
brand_models = df.groupby('brand')['name'].nunique().sort_values(ascending=False)
print(brand_models.head(10))

fig = px.bar(x=brand_models.index, y=brand_models.values,
             title="Models per Brand")
fig.show()
```
